```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
import nltk
```

```python
nltk.download('punkt')
```

```python
dataset = load_dataset("multi_news", "3.0.0")

# Convert to pandas DataFrame for easier manipulation
df = pd.DataFrame(dataset['train'])
```

```python
# Function to calculate lengths
def calculate_lengths(df):
    df['article_length'] = df['document'].apply(lambda x: len(nltk.word_tokenize(x)))
    df['summary_length'] = df['summary'].apply(lambda x: len(nltk.word_tokenize(x)))
    return df

# Calculate lengths
df = calculate_lengths(df)
```

```python
# Plotting the lengths
plt.figure(figsize=(14, 7))

plt.subplot(1, 2, 1)
plt.hist(df['article_length'], bins=50, color='blue', alpha=0.7)
plt.title('Distribution of Article Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['summary_length'], bins=50, color='green', alpha=0.7)
plt.title('Distribution of Summary Lengths')
plt.xlabel('Number of Words')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Bar graph of lengths for first 20 examples
df.head(20).plot(kind='bar', y=['article_length', 'summary_length'], figsize=(14, 7))
plt.title('Article and Summary Lengths for First 20 Examples')
plt.xlabel('Example Index')
plt.ylabel('Number of Words')
plt.show()
```
